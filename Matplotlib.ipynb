```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(
    title="Dan's Nut Butter Store",
    ylabel="Price ($)");
```

```python
stocks = {
    "Berkshire Hathaway": 407.53,
    "Apple": 184.08,
    "Microsoft": 411.60
}

fig, ax = plt.subplots()
ax.bar(stocks.keys(), stocks.values());
ax.set(
    title="S&P500",
    ylabel="Price($)",
    xlabel="Names"
);
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x / 2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.random(100));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# Plot to each different index
ax[0, 0].plot(x, x / 2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.random(100));
```

## Plotting from pandas DataFrames

```python
import pandas as pd
import numpy as np
```

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range("1/1/2020", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
# Change the format of price to number, the set didn't work for some reason
car_sales["Price"] = [int(x.replace('$', '').replace(',', '').replace('.', '')[:-2]) for x in car_sales["Price"]]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range('1/1/2020', periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=15);
```

```python
heart_disease.plot.hist(figsize=(10, 30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot is ok for simple stuff, not for this kind of advanced plots
over_50.plot(kind="scatter",
             x="age",
             y="chol", # cholesterol
             c="target"); # color by target
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6)) # in inches
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
ax.set_xlim([45, 100]) # x runs from to
```

```python
## OO method pure
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol (mg/dl)");

ax.legend(*scatter.legend_elements(), title="Has heart disease");

ax.axhline(over_50["chol"].mean(),
          linestyle="dashed")
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach - which is a max heartrate
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
# ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="dashed")

scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
# ax1.legend(*scatter1.legend_elements(), title="Has Heart Disease")
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
fig.legend(*scatter.legend_elements(), title="Has Heart Disease");
```
